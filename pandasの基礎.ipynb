# 欠損値の扱い

欠損値とは、何かしらの理由でデータが無い箇所を言います。
実データでは、実験ミスや記載漏れ、アンケートなどでは取得できなかった場合などが考えられます。
それらのデータがない箇所を、どのように取り扱うかという命題をここでは扱います。


## 2つのデータフレームをつなげた、df_new を使う

````python
df_new
````

```
先程作った、３０行のデータフレーム df_new を使いましょう。
作った時に説明しましたが、 カラム max と mean に NaNがあります。
今回のケースでは、最大値も平均値も、カラムから求めることができますが、そのような方法がわからないというケースだと想定して、欠損値の扱いを見ていきます
```

````python
df = df_new
````

```
今まで、df という変数で各種処理を行っていたので、ここからもdfが使えるように、df_newをdfにします。
このことで、最初に作った dfはデータを失います。
```

````python
df
````

```
dfを確認しています
```

````python
df.shape
````

```
行列の形も確認します
```

````python
df.tail(6)
````

```
欠損値が最後に固まっていましたので、それを見てみましょう。
列の最後を出力する tailメソッドに6行出力するように指定しています。
```


## 欠損値を、0で埋める

````python
df.fillna(0)
````

```
欠損値はデータが不明なので、計算できるように0で埋めたデータフレームを出力します。
ここでは、データは出力されますが、df自体の変更は行っていません。
```

## 欠損値がある行を削除する

````python
df.dropna()
````

```
不明なデータは今後の処理に問題になると考え、無かったものとするために行ごと削除しています。
この処置が正しいかどうかは、データの種類や欠損値の数によりますが、重要な値が抜けている場合は、列の削除を行ってしまいましょう。
このケースでも、データフレーム df に変更は与えていません。
```

## 欠損値を、カラム平均で埋める

````python
df.isnull()
````

```
データが欠損値かどうかを判定した、同じ大きさのデータフレームを出力しています
```

````python
df.notnull()
````

```
今度は反対に、欠損値の時にTrueになるメソッドで判定結果を出力しています。
```

````python
df[df['mean'].notnull()]['mean'].mean()
````

```
ここでは、欠損値でない、カラム　mean の列の平均値を求めています。
```

````python
df_fill_mean = df.fillna({'mean': df[df['mean'].notnull()]['mean'].mean()})
````

```
欠損値でないカラム mean の平均値を、欠損値に入れています。
ここでは、新たにデータフレーム df_fill_mean を生成しています。
```

````python
df_fill_mean
````

```
新たに生成したデータフレームを見てみましょう。
カラム mean に欠損値がなくなりました

欠損値の処理には、前後の値で補間する方法も提供さています。
```

## 深いコピーと浅いコピー

データの加工を行う上で、元のデータを確保したまま実行したい場合があります。
その際には、深いコピー（ディープコピー）を行うことになります。

浅いコピーは、通常の参照と変わりません。
