## NumPyで行列作成

````python
import numpy as np
````

```
NumPyをnpという名前でインポートしています
NumPyは、Cで実装された高効率な配列や行列を扱うオブジェクトを提供しています。
pandasのオブジェクトの内部はNumPyの配列オブジェクトが利用されており、密接な関係があります。
Pythonには、NumPyがあるからこそ、データ系の処理に向いているとも言われています。
```

````python
np.arange(100).reshape(25, 4)
````

```
NumPyのarangeで、0から99までの連続した整数の配列を作り、それを、25行4列の2次元配列を生成しました。
```

# データフレームの作成

````python
df = pd.DataFrame(np.arange(100).reshape(25, 4), columns=list("abcd"))
df
````

```
上記のNumPyで生成した25行4列の二次元配列をpandasのデータフレームに取り込んでいます。
その際に、カラム名として、a, b, c, d と名前をつけています。
```
